```python
def make_id(row):   
    if row['ensembl'] == 'NaN' and  row['ensemble.gene'] == 'NaN': # 1) If both the value of 'ensembl' and 'ensembl.gene' is 'NaN', then output is "Not found".
        return 'Not Found'
    elif row['ensembl'] == 'NaN':                                  # 2) If the value of 'ensembl' is 'NaN', then just print the value of 'ensembl.gene'
        return row['ensembl.gene']
    else:                                                          # 3) (otherwise) If the value of 'ensembl.gene' is 'NaN', then print first part of the value of 'ensembl' 
        return row['ensembl']


s['id'] = s.apply(lambda x: make_id(x), axis=1)
print(s['id'])
```
